```python
#1
import numpy as np
import matplotlib.pyplot as plt

a = np.loadtxt("http://www.scipy-lectures.org/_downloads/populations.txt",  dtype = 'int', skiprows = 1, delimiter = "\t")
plt.figure(1, figsize=(8, 5))
plt.plot(a[:, 0], a[:, 1], label = "Hares", color = 'darkgrey')
plt.plot(a[:, 0], a[:, 2], label = "Lynxes", color = 'brown')
plt.plot(a[:, 0], a[:, 3], label = "Carrots", color = 'orangered')

plt.legend()
plt.xlabel("Years")
plt.ylabel("Population")
plt.show()
```

```python
#2
totpop = a[:, 1:3].sum(axis = 1)
plt.figure(1, figsize=(8, 5))
plt.bar(a[:, 0], a[:, 1]/totpop, bottom = a[:, 2]/totpop, label = "Hares", color = 'darkgrey')
plt.bar(a[:, 0], a[:, 2]/totpop, label = "Lynxes", color = 'brown')
plt.legend()
plt.show()
```

```python
#3
print ("Hares: mean =", np.mean(a[:, 1]), ", std =", np.std(a[:, 1]))
print ("Lynxes: mean =", np.mean(a[:, 2]), ", std =", np.std(a[:, 2]))
print ("Carrots: mean =", np.mean(a[:, 3]), ", std =", np.std(a[:, 3]))

plt.figure(1, figsize=(15, 5))
plt.subplot (131)
plt.hist(a[:, 1], color = 'darkgrey')
plt.axvline(np.mean(a[:, 1]), color = 'black')
plt.title("Hares")

plt.subplot (132)
plt.hist(a[:, 2], color = 'brown')
plt.axvline(np.mean(a[:, 2]), color = 'black')
plt.title("Lynxes")
 
plt.subplot (133)
plt.hist(a[:, 3], color = 'orangered')
plt.axvline(np.mean(a[:, 3]), color = 'black')
plt.title("Carrots")
plt.show()
```

```python
#4
print ("Year of the largest population")
print ("Hares :", a[:, 0] [np.argmax(a[:, 1])])
print ("Lynxes :", a[:, 0] [np.argmax(a[:, 2])])
print ("Carrots :", a[:, 0] [np.argmax(a[:, 3])])

plt.figure(1, figsize=(8, 5))
plt.plot(a[:, 0], a[:, 1], a[:, 0] [np.argmax(a[:, 1])], max(a[:, 1]), 'o', color = 'darkgrey')
plt.plot(a[:, 0], a[:, 2], a[:, 0] [np.argmax(a[:, 2])], max(a[:, 2]), 'o', color = 'brown')
plt.plot(a[:, 0], a[:, 3], a[:, 0] [np.argmax(a[:, 3])], max(a[:, 3]), 'o', color = 'orangered')

plt.xlabel("Years")
plt.ylabel("Population")
plt.legend (("Hares", "Largest population", "Lynxes", "Largest population", "Carrots", "Largest population"))
plt.show()
```

```python
#5
result = (np.where(a[:, 1] > a[:, 2], 'H', 'L')).tolist()
print (result)
plt.pie ([result.count('H'), result.count('L')], labels = ["'H'", "'L'"], colors=["darkgrey","brown"], autopct='%i%%')
plt.show()
```

```python
#6
H = np.array(a[:, 0] [a[:, 1] > 50000])
L = np.array(a[:, 0] [a[:, 2] > 50000])
C = np.array(a[:, 0] [a[:, 3] > 50000])
print("Years when populations of Hares is above 50000: " , H)
print("Years when populations of Lynxes is above 50000: " , L)
print("Years when populations of Carrots is above 50000: " , C)

plt.figure(1, figsize=(8, 5))
plt.plot(a[:, 0], a[:, 1], H, a[:, 1] [a[:, 1] > 50000], 'o', color = 'darkgrey')
plt.plot(a[:, 0], a[:, 2], L, a[:, 2] [a[:, 2] > 50000], 'o', color = 'brown')
plt.plot(a[:, 0], a[:, 3], C, a[:, 2] [a[:, 3] > 50000], 'o', color = 'orangered')

plt.legend (("Hares", "Above 50000", "Lynxes", "Above 50000", "Carrots", "Above 50000"))
plt.xlabel("Years")
plt.ylabel("Population")
plt.show()
```

```python
#7
H = np.sort(a[:, 1])
L = np.sort(a[:, 2])
C = np.sort(a[:, 3])
print("Top 2 years when populations of Hares is the lowest: ", a[:, 0] [a[:, 1] == H[0]], a[:, 0] [a[:, 1] == H[1]])
print("Top 2 years when populations of Lynxes is the lowest: ", a[:, 0] [a[:, 2] == L[0]], a[:, 0] [a[:, 2] == L[1]])
print("Top 2 years when populations of Carrots is the lowest: ", a[:, 0] [a[:, 3] == C[0]], a[:, 0] [a[:, 3] == C[1]])
```

```python
#8
plt.figure(1, figsize=(8, 5))
plt.plot(a[:, 0], a[:, 2], label = "The number of lynxes", color = 'brown')
plt.plot(a[:, 0], [0] + list((a[1:, 1] - a[:len(a)-1, 1])) , label = "The change in hare population", color = 'darkgrey')

plt.xlabel("Years")
plt.ylabel("Population")
plt.legend()
plt.show()
print("Correlation coefficient =", np.corrcoef(a[:, 2], [0] + list((a[1:, 1] - a[:len(a)-1, 1])))[0, 1])
```
