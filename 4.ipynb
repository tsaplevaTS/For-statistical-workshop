### Import pandas, numpy and matplotlib

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

### Import the dataset from this [address]('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')

````python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
df = pd.read_csv(url, sep = '\t')
````

### Show the first 10 entries

````python
df.head(10)
````

### What is the number of observations in the dataset?

````python
df.shape[0]
````

### What is the number of columns in the dataset?

````python
df.shape[1]
````

### Which was the most-ordered item and how many times it was ordered?

````python
df['item_name'].value_counts()[:1]
````

### How many different items are sold?

````python
len(df.groupby('item_name'))
````

### How many items were orderd in total?

````python
df['quantity'].sum()
````

### Turn the item price into a float

````python
df['item_price'] = df['item_price'].map(lambda x: x.lstrip('$').rstrip(' '))
df['item_price']=df['item_price'].apply(pd.to_numeric)
df.head(10)
````

### What is the price of each item?

````python
df[['item_name', 'item_price']].drop_duplicates()
````

### How much was the revenue for the period in the dataset?

````python
df['revenue']= (df['quantity'])*(df['item_price'])
df['revenue'].sum()
````

### How many orders were made in the period?

````python
len(df.groupby('order_id'))
````

### What is the average revenue amount per order?

````python
df['revenue'].sum()/len(df.groupby('order_id'))
````

### Merge the table ```dfa``` that contains order date and order id with ```df``` and plot daily revenue 

````python
#Т.к. дат меньше, чем количество заказов, то таблица урезается
dfa = pd.DataFrame({"date": np.random.choice(pd.date_range(start="20170101", end="20170201"), size=1800),
                   "order_id": np.arange(1800)})
df = pd.merge(dfa, df)
df
````

````python
df.groupby('date').sum()['revenue']
````

````python
plt.figure(1, figsize=(10,8))
df.groupby('date').sum()['revenue'].plot(x = 'date',y = 'revenue', color = 'black')
plt.xlabel("Iem_name")
plt.ylabel("Revenue")
````

### Create a histogram of the top 5 items bought

````python
df1 = df['item_name'].value_counts()[:5]
plt.figure(1, figsize=(10,8))
plt.bar (df1.index, df1.values,  color = 'black')
plt.xlabel("Item name")
plt.ylabel("Quantity")
````

### How many times people orderd more than one Canned Soda?

### Create a scatterplot with the number of items orderered per order price

````python
plt.scatter(df.groupby('order_id').sum()['quantity'], df.groupby('order_id').sum()['revenue'], color='black', s = 10, linewidth = 1)
````
