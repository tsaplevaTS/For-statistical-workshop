```python
import numpy as np
import math
y = np.array([7, 7, 1, 3, 2, 1, 5, 1, 4, 0, 2, 3, 3, 1, 2, 2, 3, 1, 4, 0, 8, 3, 5, 3, 0, 0, 7, 1, 1, 3, 1, 3, 2, 4, 1, 6, 2, 2, 4, 1, 3, 1, 1, 2, 7, 3, 3, 2, 2, 2])
M = 10
N = len(y) 
def prob(y_i,m_s):
    return math.factorial(M)/math.factorial(y_i)/math.factorial(M - y_i)*(m_s)**y_i*(1 - m_s)**(M - y_i)
def r_coef(i,k):
    return prob(y[i],m[t - 1][k])*pi[t - 1][k]/(prob(y[i],m[t - 1][0])*pi[t - 1][0] + prob(y[i],m[t - 1][1])*pi[t - 1][1])

pi = []
m = []
pi.append([0.7,0.3])
m.append([0.6,0.4])
t = 1
r = np.zeros((N,2))
while 1:
    for i in range(N):
        for j in range(2):
            r[i][j] = r_coef(i, j) #(1)
    pi.append(np.sum(r, axis = 0)/N) #(2)
    m.append([np.sum(r[:, 0]*y)/np.sum(r[:, 0])/M, np.sum(r[:, 1]*y)/np.sum(r[:, 1])/M]) #(3)
    if (np.abs(m[t][0] - m[t - 1][0]) < 1e-10): #проверка значимости изменения оценки параметров
        print (m)
        break
    t += 1
print (t)
```
